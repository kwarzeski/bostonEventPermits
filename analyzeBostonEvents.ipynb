```python
# summon librarys
import json
from datetime import datetime, timedelta
import urllib.request
```

```python
# pull data for special event liscense applications - see analyze boston website https://data.boston.gov/dataset/special-event-license-applications for details
url = 'https://data.boston.gov/datastore/odata3.0/fb05bfda-a57e-4d97-948b-4f1949ba3c8a?$format=json'
fileobj = urllib.request.urlopen(url)
response_dict = json.loads(fileobj.read())
# print(response_dict['value'])
```

```python
# Get events from response
events = response_dict['value']
# events[0]
```

```python
# Function to turn string to datetime. If there is no time, add a time. If void, return 2000-01-01
def timeSort(timeString):
  if timeString:
    try:
      return datetime.strptime(timeString, '%Y-%m-%d %H:%M:%S')
    except:
      return datetime.strptime(timeString, '%Y-%m-%d')
  else:
    return datetime(2000, 1, 1)
```

```python
# Convert values from string
for i, element in enumerate(events):
  events[i]["Start Date and Time"] = timeSort(element["Start Date and Time"])
  events[i]["End Date and Time"] = timeSort(element["End Date and Time"])
```

```python
# Drop all events that are in the past
# Includes events today
now = datetime.now()
upcoming_Events = [x for x in events if x["Start Date and Time"] > now - timedelta(days=1)]
```

```python
now = datetime.now()

# Print for each event-
# date, event name, status, if approved, location
for event in upcoming_Events:
  # get events with a date/time and a name
  if event["Start Date and Time"] and event["App Name"] and event["Status"] == "Open":
    d = event["Start Date and Time"]
    # get events in the next 20 days
    if d < now + timedelta(days=20):
      print(d.strftime("%b %d %Y"))
      if not event["Start Date and Time"].hour == 0:
        print(event["Start Date and Time"].strftime("%H:%M"))
      print(event["App Name"])
      print(event["Status"])
      # Print a google search link
      print("https://www.google.com/search?q="+event["App Name"].strip().replace(' ', "+")+"+"+event["City"].strip().replace(' ', "+"))
      if event['Issued Date']:
        print("Approved")
      print(event["City"])
      print(event["Location Description"])
      print("---")
```
